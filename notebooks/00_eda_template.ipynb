# Analyze

## Q1 — [Does X relate to Y?]

```python
# Q1 — replace column names
correlation = df['[col_x]'].corr(df['[col_y]'])
print(f'Correlation: {correlation:.3f}')

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='[col_x]', y='[col_y]', alpha=0.4)
sns.regplot(data=df, x='[col_x]', y='[col_y]', scatter=False, color='red')
plt.title('[X] vs. [Y]')
plt.xlabel('[X label]')
plt.ylabel('[Y label]')
plt.tight_layout()
plt.savefig(visual('q1_x_vs_y.png'), dpi=150)
plt.show()
```

**Q1 Note:** Pearson's r interpretation

- Near 0 → No relationship
- 0.3 to 0.5 → Moderate
- 0.7 to 1.0 → Strong

## Q2 — [Which category/segment performs best?]

```python
# Q2 — replace column names
category_metric = (
    df.groupby('[category_col]')['[metric_col]']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=category_metric, x='[metric_col]', y='[category_col]',
    palette='viridis', hue='[category_col]', legend=False,
)
plt.title('[Metric] by [Category]')
plt.xlabel('[Metric label]')
plt.ylabel('[Category label]')
plt.tight_layout()
plt.savefig(visual('q2_metric_by_category.png'), dpi=150)
plt.show()
```
